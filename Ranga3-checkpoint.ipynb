# Knn Algorithm
```
1.Read the csv file
2.Seperate the feature values and target
3. split the data into training and testing
4.import required libraries
5. fit the model
6.predict the values
7. score
```

````python
# 1.read the csv file
from sklearn.datasets import load_iris
````

````python
iris = load_iris()
iris
````

````python
x_actual  = iris.data
x_actual
````

````python
y_actual = iris.target
y_actual
````

````python
import pandas as pd
````

````python
y=pd.get_dummies(iris.target)
y_actual = y[0]
````

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_actual,y_actual,test_size=0.3,random_state=1)
````

````python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier()
````

````python
knn.fit(x_train,y_train)
````

````python
y_pred = knn.predict(x_test)
y_pred
````

````python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
````

````python
knn1= KNeighborsClassifier(n_neighbors=1)
````

````python
knn1.fit(x_train,y_train)
````

````python
y_pred=knn1.predict(x_test)
````

````python
accuracy_score(y_test,y_pred)
````

````python
knn2=KNeighborsClassifier(n_neighbors=2)
````

````python
knn2.fit(x_train,y_train)
````

````python
y_pred1=knn2.predict(x_test)
````

````python
accuracy_score(y_test,y_pred1)
````

````python
import numpy as np
import pandas as pd

data=pd.read_csv("bird.csv")
data
````

````python
Type=pd.get_dummies(data['type'],drop_first=True)
Type
````

````python
data1=pd.concat([data,Type],axis=1)
data1
````

````python
data1=data1.drop('type',axis=1)
````

````python
data1
````

````python
x=data1.drop(['R','SO','SW','T','W'],axis=1)
x
````

````python
y=data1[['R','SO','SW','T','W']]
y
````

````python
x=x.drop('id',axis=1)
x
````

````python
x.isnull().sum()
````

````python
x = x.replace(np.nan,0)
````

````python
x.isnull().sum()
````

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
````

````python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()

knn.fit(x_train,y_train)

y_pred=knn.predict(x_test)
````
