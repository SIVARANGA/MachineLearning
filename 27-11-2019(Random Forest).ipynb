```python
import pandas as pd
```

```python
data=pd.read_csv("diabetes.csv")
data.head()
```

```python
x=data.drop('Outcome',axis=1)
y=data['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()

rfc.fit(x_train,y_train)

rfc_predict=rfc.predict(x_test)
```

```python
# Evaluating performance
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report,confusion_matrix
```

```python
rfc_cv_score=cross_val_score(rfc,x,y,cv=10,scoring='roc_auc')

rfc_cv_score
```

```python
print("confussion_matrix",confusion_matrix(y_test,rfc_predict))
print("classification_report",classification_report(y_test,rfc_predict))
print("rfc_cv_score",rfc_cv_score)
print("rfc_cv_score.mean",rfc_cv_score.mean())
```

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
n_estimators=[int(x) for x in np.linspace(start=200,stop=2000,num=11)]

max_features=['auto','sqrt']

max_depth=[int(x) for x in np.linspace(100,500,num=11)]

max_depth.append(None)

random_grid={'n_estimators':n_estimators,'max_features':max_features,"max_depth":max_depth}

rfc_random=RandomizedSearchCV(estimator=rfc,param_distributions=random_grid,n_iter=100,cv=3,verbose=2,
                              random_state=42,n_jobs=-1)

rfc_random.fit(x_train,y_train)

print(rfc_random.best_params_)
```

```python
# Evaluating performance
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report,confusion_matrix
```

```python
rfc=RandomForestClassifier(n_estimators=200,max_depth=140,max_features="sqrt")

rfc.fit(x_train,y_train)

rfc_predict=rfc.predict(x_test)
rfc_cv_score=cross_val_score(rfc,x,y,cv=10,scoring='roc_auc')

print(confusion_matrix(y_test,rfc_predict))
print("\n")
print(classification_report(y_test,rfc_predict))
print("\n")
print(rfc_cv_score)
print("\n")
print(rfc_cv_score.mean())
```
